Now add a binary feature to the DataFrame – `Customer service calls > 3`. And name it 'Many service calls' after that see how it relates to churn. 

```python
df['Many_service_calls'] = (df['Customer service calls'] > 3)

pd.crosstab(df['Many_service_calls'], df['Churn'], margins=True)
```

```python
#your code for plot

sns.countplot(x='Many_service_calls', hue='Churn', data=df)
```

Construct another contingency table that relates *Churn* with both *International plan* and freshly created *Many_service_calls*.

```python
pd.crosstab(df['Many_service_calls'] & df['International plan'],df['Churn'])
```

Give overall inference and which variable can used as best predictor for churn.

From the analysis, the customer is not loyal in the case where Customer service calls greater than 3.

We might expect an accuracy of 86.0 percentage which is the share of loyal customer in the data.

There is a large proportion of churned customers respect to Many service calls.

Therefore we conclude that Many service calls variable is the best predictor for churn.
